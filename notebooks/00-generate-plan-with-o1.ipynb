```python
import os
import json
import openai
from openai import OpenAI
# from llama_index.llms.openai import OpenAI
from dotenv import load_dotenv
```

```python
load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
openai.api_key = OPENAI_API_KEY
```

```python
# Create a client instance
client = openai.OpenAI(api_key=OPENAI_API_KEY)
```

```python
from tabulate import tabulate
models = client.models.list()

# Extract relevant details
table = [[model.id, model.created, model.owned_by] for model in models.data]

# Print in table format
print(tabulate(table, headers=["Model ID", "Created", "Owner"], tablefmt="grid"))
```

```python
# Load latest plan prompt
prompt = json.loads(open("../misc/o1-plan-generation-prompt.json").read())

response = client.chat.completions.create(
    model="o1-preview-2024-09-12", # o1-mini for now just for testing
    messages=[{"role": "user", "content": str(prompt)}],
)

print(response.choices[0].message.content)  # Access the response message
```
